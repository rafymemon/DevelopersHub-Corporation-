# Visualizing the dataset using

## a. Scatterplot to show the relationship between features
## b. histograms to show the value distributions
## c. boxplots to identify the outliers

```python
# Setting the style for better visualizations
sns.set_style("whitegrid")
```

```python
# a. Scatterplot to show the realtionship between the features

# scatterplot between sepal_length and sepal_width
plt.figure(figsize = (12, 8))
plt.title('Sepal Length vs Sepal Width by Species')

sns.scatterplot(x = iris['sepal_length'], y = iris['sepal_width'], hue = iris['species'])
plt.show()
```

```python
plt.figure(figsize = (12, 8))

# scatterplot between petal_length and petal_width
plt.title('Petal Length vs Petal Width by Species')

sns.scatterplot(x = iris['petal_length'], y = iris['petal_width'], hue = iris['species'])
plt.show()
```

```python
# b. Using histograms to show value distributions of different features
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Distribution of Iris Features')

sns.histplot( x=iris['sepal_length'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length Distribution')

sns.histplot(data=iris, x='sepal_width', kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width Distribution')

sns.histplot(data=iris, x='petal_length', kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Petal Length Distribution')

sns.histplot(data=iris, x='petal_width', kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Petal Width Distribution')

plt.tight_layout() # used to make the titles and layouts in proper order
plt.show()
```

```python
# c. boxplot to identify outliers
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=iris, x='species', y=column)
    plt.title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```
